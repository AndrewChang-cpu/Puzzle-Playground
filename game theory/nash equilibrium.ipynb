```python
import nashpy as nash
import numpy as np

A = np.array([[5,1,4,4], [2,0,3,1], [3,0,4,1], [5,2,0,0], [1,1,6,4]])
B = np.array([[1,2,0,4,6], [0,0,5,0,6], [3,5,3,2,0], [0,5,5,5,5]])
rps = nash.Game(A, B.T)
print(rps)
```

```python
list(rps.support_enumeration())
```

```python
from scipy.optimize import linprog

A = -np.array([[1,0,3,0,-1], [2,0,5,5,-1], [0,5,3,5,-1], [4,0,2,5,-1], [6,6,0,5,-1], [-1,-1,-1,-1,0], [1,1,1,1,0]])
b = [0, 0, 0, 0, 0, -1, 1]
c = [0, 0, 0, 0, 1]
linprog(c, A_ub=A, b_ub=b)
```
