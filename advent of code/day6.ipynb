```python
with open("day6_input.txt", "r") as f:
    lines = list(f.readlines())
    grid = [list(map(int, x.split())) for x in lines[:-1]]
    ops = lines[-1].strip().split()
    res = 0
    for col in range(len(grid[0])):
        if ops[col] == '+':
            res += sum(grid[row][col] for row in range(len(grid)))
        else:
            curr = 1
            for row in range(len(grid)):
                curr *= grid[row][col]
            res += curr
    print(res)
```

```python
with open("day6_input.txt", "r") as f:
    text = list(f.readlines())
#     text = """123 328  51 64 
#  45 64  387 23 
#   6 98  215 314
# *   +   *   +  """.splitlines()
    ops = text[-1].strip().split()
    grid = text[:-1]
    print(grid, [len(x) for x in grid], len(grid))
    col_idx = 0
    res = 0
    for op in ops:
        total = 0 if op == '+' else 1
        while col_idx < len(grid[0]) - 1:
            populated = False
            curr = ''
            for row in range(len(grid)):
                if grid[row][col_idx] != ' ':
                    populated = True
                    curr += grid[row][col_idx]
            col_idx += 1
            if not populated:
                break
            
            val = int(curr)
            if op == '+':
                total += val
            else:
                total *= val
            print(curr, end=', ')
        print(total)
        res += total
    print(res)
```
