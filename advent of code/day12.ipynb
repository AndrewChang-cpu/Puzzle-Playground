```python
with open("day12_input.txt") as f:
    text = f.read().strip()
s1, s2, s3, s4, s5, s6, queries = text.split('\n\n')
shapes = [s.split(':\n')[1] for s in [s1, s2, s3, s4, s5, s6]]
queries = queries.split('\n')
new_queries = []
for q in queries:
    dim, n = q.split(': ')
    h, w = map(int, dim.split('x'))
    qty = list(map(int, n.split(' ')))
    new_queries.append((h, w, qty))
queries = new_queries
```

```python
invalid = 0
potential = 0
trivial = 0
potential_queries = []
for i, q in enumerate(queries):
    h, w, qty = q
    squares_needed = 0
    for j in range(len(shapes)):
        squares_needed += qty[j] * shapes[j].count('#')
    if squares_needed > h * w:
        invalid += 1
    elif sum(qty) <= (h // 3) * (w // 3):
        trivial += 1
    else:
        potential += 1
        potential_queries.append(q)
print(invalid)
print(trivial)
print(potential)
potential_queries[:5]
```
