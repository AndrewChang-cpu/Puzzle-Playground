```python
def base():
    with open("day1_input.txt", "r") as f:
        res = 0
        pos = 50
    #     f = """L68
    # L30
    # R48
    # L5
    # R60
    # L55
    # L1
    # L99
    # R14
    # L82""".split()
        # f = ['L50', 'L100', 'L99']
        for line in f:
            r = line.strip()
            if r[0] == 'R':
                for _ in range(int(r[1:])):
                    pos += 1
                    pos %= 100
                    res += pos == 0
            else:
                for _ in range(int(r[1:])):
                    pos -= 1
                    pos %= 100
                    res += pos == 0
            yield res, pos
        # print(res)
```

```python
with open("day1_input.txt", "r") as f:
    iter_base = base()
    iter_efficient = efficient()
    # print(len(list(iter_base)), len(list(iter_efficient)))
    for _ in range(4186):
        x, px = next(iter_base)
        y, py = next(iter_efficient)
        line = next(f)
        print(line.strip(), x, y, px, py)
        if x != y:
            print("Mismatch!")
            break
```
