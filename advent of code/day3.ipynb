```python
with open('day3_input.txt', 'r') as f:
    jolts = 0
    for line in f:
        curr = line.strip()
        max1 = max(curr[:-1])
        max1_idx = curr.find(max1)
        max2 = max(curr[max1_idx + 1:])
        jolts += int(max1) * 10 + int(max2)
    print(jolts)
```

```python
with open('day3_input.txt', 'r') as f:
    jolts = 0
    for line in f:
        curr = line.strip()
        start = 0
        for i in range(12):
            print(curr[start:])
            max1 = max(curr[start:len(curr) - 11 + i])
            max1_idx = curr.find(max1, start)
            start = max1_idx + 1
            jolts += int(max1) * 10 ** (11 - i)
        print()
    print(jolts)
```
