```python
from collections import deque, defaultdict

adj = defaultdict(list)
with open('day11_input.txt') as f:
    for line in f:
        parent, children = line.split(': ')
        children = line.split()
        adj[parent] = children

q = deque(['you'])
paths = defaultdict(int, you=1)
paths_to_out = 0
while q:
    next_paths = defaultdict(int)
    for _ in range(len(q)):
        node = q.popleft()
        if node == 'out':
            paths_to_out += paths['out']
        for nbr in adj[node]:
            if nbr not in next_paths:
                q.append(nbr)
            next_paths[nbr] += paths[node]
    paths = next_paths
print(paths_to_out)
```

```python
from functools import cache

@cache
def find_paths_to_out(node):
    if node == 'out':
        return 1, 0, 0, 0
    
    neither = dac = fft = both = 0
    for nbr in adj[node]:
        n, d, f, b = find_paths_to_out(nbr)
        if node == 'dac':
            dac += n + d
            both += f + b
        elif node == 'fft':
            fft += n + f
            both += d + b
        else:
            neither += n
            dac += d
            fft += f
            both += b
    
    return neither, dac, fft, both

print(find_paths_to_out('svr'))
```
